#### Measure of central tendence 

```python

sale_price = df_train['SalePrice']

mean_val   = sale_price.mean()
median_val = sale_price.median()
mode_val   = sale_price.mode().iloc[0]  

print('Measure of central tendence')
print('')
print('rata-rata: ',mean_val)
print('median: ',median_val)
print('mode: ',mode_val)

```

#### Measure of Spread

```python
#Measure of Spread


std_val  = sale_price.std()
var_val = sale_price.var()
Q1_val = sale_price.quantile(0.25)
Q3_val= sale_price.quantile(0.75)
IQR_val = Q3_val-Q1_val
print('Measure of Spread')
print('')
print('Standar deviasi: ',round(std_val,2))
print('varian: ',round(var_val,2))
print('Inter Quartile Range (IQR):',IQR_val) #Q3-Q1
```

```python
"""
Boxplot ‑ SalePrice  (dengan anotasi Median, Q1, Q3, IQR, Whisker & Outlier)
---------------------------------------------------------------------------
• Pastikan file  saleprice.csv  berada di folder kerja yang sama,
  atau ganti path di pd.read_csv().
"""

import pandas as pd
import matplotlib.pyplot as plt

# 1) BACA DATA ---------------------------------------------------------------
df = pd.read_csv("train.csv")          # ganti path jika perlu
data = df["SalePrice"]                     # kolom numerik yang akan diplot

# 2) HITUNG LIMA ANGKA PENTING ----------------------------------------------
Q1  = data.quantile(0.25)
Q2  = data.median()
Q3  = data.quantile(0.75)
IQR = Q3 - Q1
lower_whisker = max(data.min(), Q1 - 1.5 * IQR)    # aturan Tukey 1.5×IQR
upper_whisker = min(data.max(), Q3 + 1.5 * IQR)

# 3) DETEKSI OUTLIER (opsional) ---------------------------------------------
outliers = data[(data < lower_whisker) | (data > upper_whisker)]

# 4) BUAT BOXPLOT ------------------------------------------------------------
plt.figure(figsize=(8, 6))
plt.boxplot(
    data,
    vert=True,
    patch_artist=True,
    widths=0.4,
    boxprops   = dict(facecolor="lightblue", color="blue"),
    medianprops= dict(color="red"),
    whiskerprops=dict(color="black"),
    capprops   = dict(color="black"),
    flierprops = dict(marker="o", color="orange", alpha=0.6)
)

# 5) ANOTASI -----------------------------------------------------------------
plt.text(1.1, Q2, f"Median:\n{Q2:,.0f}",   va="center", fontsize=9, color="red")
plt.text(1.1, Q1, f"Q1:\n{Q1:,.0f}",       va="center", fontsize=9, color="blue")
plt.text(1.1, Q3, f"Q3:\n{Q3:,.0f}",       va="center", fontsize=9, color="blue")
plt.text(0.80, lower_whisker,
         f"Lower Whisker:\n{lower_whisker:,.0f}", ha="right", fontsize=8)
plt.text(0.80, upper_whisker,
         f"Upper Whisker:\n{upper_whisker:,.0f}", ha="right", fontsize=8)
plt.text(1.25, Q3 + 0.10 * IQR, f"IQR:\n{IQR:,.0f}", fontsize=9, color="purple")

# 6) TATA LETAK --------------------------------------------------------------
plt.title("Annotated Boxplot of SalePrice")
plt.ylabel("SalePrice (USD)")
plt.xticks([])                              # sumbu‑X tidak dipakai (hanya 1 variabel)
plt.tight_layout()
plt.show()
```
