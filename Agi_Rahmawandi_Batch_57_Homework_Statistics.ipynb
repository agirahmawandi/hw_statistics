#### Measure of central tendence 

```python

sale_price = df_train['SalePrice']


mean_val   = sale_price.mean()
median_val = sale_price.median()
min_val = sale_price.min()
max_val = sale_price.max()
mode_val   = sale_price.mode().iloc[0] 
Q1_val = sale_price.quantile(0.25)
Q3_val= sale_price.quantile(0.75)


print('Measure of central tendence')
print('')
print('rata-rata: ',mean_val)
print('median: ',median_val)
print('mode: ',mode_val)
print('min: ',min_val)
print('max: ',max_val)
print('Q1',Q1_val)
print('Q3',Q3_val)
```

#### Measure of Spread

```python
#Measure of Spread


std_val  = sale_price.std()
var_val = sale_price.var()
IQR_val = Q3_val-Q1_val
lower_whisker =  Q1_val - (1.5 * IQR_val)   # aturan Tukey 1.5×IQR
upper_whisker = Q3_val + (1.5 * IQR_val)
print('Measure of Spread')
print('')
print('Standar deviasi: ',round(std_val,2))
print('varian: ',round(var_val,2))
print('Inter Quartile Range (IQR):',IQR_val) #Q3-Q1
print('lower whisker :',lower_whisker)
print('upper whisker',upper_whisker)
```

### interpretasi
- Berdasarkan data transaksi harga rata‑rata rumah adalah 214K, sedangkan median‑nya 188K Perbedaan mean > median menunjukkan distribusi miring ke kanan — ada sejumlah properti premium yang menaikkan rata‑rata. Modus di kisaran  175K, menandakan titik harga paling diminati pembeli.

- Simpangan baku sebesar 79K dan IQR 92K mengindikasikan variabilitas harga cukup lebar; manajemen perlu menetapkan segmentasi pasar agar strategi pricing lebih presisi. Range keseluruhan (35K – 755K) menegaskan adanya segmen properti high‑end yang, meski volumenya kecil, tetap signifikan terhadap total penjualan.”
