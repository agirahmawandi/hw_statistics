#### Measure of central tendence 

```python

sales_price = df_train['SalePrice']

print('Measure of central tendence')
print('')
print('rata-rata: ',sales_price.mean())
print('median: ',sales_price.median())
print('mode: ',sales_price.mode()[0])

```

#### Measure of Spread

```python
#Measure of Spread
print('Measure of Spread')
print('')
print('Standar deviasi: ',round(sales_price.std(),2))
print('varian: ',round(sales_price.var(),2))
print('Inter Quartile Range (IQR):',sales_price.quantile(0.75)-sales_price.quantile(0.25)) #Q3-Q1
```

```python
#BOXPLOT
plt.figure(figsize = (7,2))
sns.boxplot(x='SalePrice',data = df_train)
plt.title('Boxplot Sale Price')
plt.tight_layout()
plt.show()
```

```python

# HISTOGRAM + KDE + MEAN, MEDIAN, MODE, ±1 SD   

df = pd.read_csv("train.csv")        
sale_prices = df["SalePrice"]

# HITUNG STATISTIK
mean_val   = sale_prices.mean()
median_val = sale_prices.median()
mode_val   = sale_prices.mode().iloc[0]   # ambil modus pertama jika >1
std_val    = sale_prices.std()

# GRAFIK HISTOGRAM + KDE
plt.figure(figsize=(10, 5))
sns.histplot(
    sale_prices,
    kde=True,
    bins=30,
    color="skyblue",
    edgecolor="black"
)

# GARIS MEAN, MEDIAN, MODE, ±1 SD
plt.axvline(mean_val,   color="blue",  ls="--", lw=2, label=f"Mean   = {mean_val:,.0f}")
plt.axvline(median_val, color="green", ls=":",  lw=2, label=f"Median = {median_val:,.0f}")
#plt.axvline(mode_val,   color="purple",ls="dashdot", lw=2, label=f"Mode   = {mode_val:,.0f}")

plt.axvline(mean_val - std_val, color="red", ls="-.", lw=2,
            label=f"-1 SD  = {mean_val - std_val:,.0f}")
plt.axvline(mean_val + std_val, color="red", ls="-.", lw=2,
            label=f"+1 SD  = {mean_val + std_val:,.0f}")

# ARSIRAN ±1 SD
plt.axvspan(mean_val - std_val, mean_val + std_val,
            color="red", alpha=0.10, label="±1 SD range")

# TATA LETAK
plt.title("Distribusi SalePrice dengan Mean, Median, Mode & ±1 SD")
plt.xlabel("SalePrice (USD)")
plt.ylabel("Frekuensi")
plt.legend()
plt.tight_layout()
plt.show()
```
