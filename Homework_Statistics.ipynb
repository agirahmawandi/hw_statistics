```python
#+Std
mean_val = df_train['SalePrice'].mean()
std_val  = df_train['SalePrice'].std()

# ---------------------------------
# 2. Histogram + mean, median, ±1SD
# ---------------------------------
plt.figure(figsize=(6, 4))

plt.hist(df_train['SalePrice'], bins=6)

# Garis pusat
plt.axvline(mean_val, ls='--', lw=1.5, label=f'Mean ({mean_val:.1f})')
plt.axvline(df_train['SalePrice'].median(), ls=':', lw=1.5,
            label=f'Median ({df_train['SalePrice'].median():.1f})')

# Garis ±1 SD
plt.axvline(mean_val - std_val, ls='-.', lw=1.2,
            label=f'-1 SD ({mean_val - std_val:.1f})')
plt.axvline(mean_val + std_val, ls='-.', lw=1.2,
            label=f'+1 SD ({mean_val + std_val:.1f})')

plt.title('Distribusi Total Penjualan')
plt.xlabel('Total Penjualan (ribu rupiah)')
plt.ylabel('Frekuensi')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ──────────────────────────────────────────────────────────────
# HISTOGRAM + KDE + MEAN, MEDIAN, MODE, ±1 SD   |  
# ──────────────────────────────────────────────────────────────
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1 ▸  BACA DATA
# ----------------------------------------------------------------
df = pd.read_csv("saleprice.csv")        # ubah path sesuai lokasi file
sale_prices = df["SalePrice"]

# 2 ▸  HITUNG STATISTIK
# ----------------------------------------------------------------
mean_val   = sale_prices.mean()
median_val = sale_prices.median()
mode_val   = sale_prices.mode().iloc[0]   # ambil modus pertama jika >1
std_val    = sale_prices.std()

# 3 ▸  GRAFIK HISTOGRAM + KDE
# ----------------------------------------------------------------
plt.figure(figsize=(10, 5))
sns.histplot(
    sale_prices,
    kde=True,
    bins=30,
    color="skyblue",
    edgecolor="black"
)

# 4 ▸  GARIS MEAN, MEDIAN, MODE, ±1 SD
# ----------------------------------------------------------------
plt.axvline(mean_val,   color="blue",  ls="--", lw=2, label=f"Mean   = {mean_val:,.0f}")
plt.axvline(median_val, color="green", ls=":",  lw=2, label=f"Median = {median_val:,.0f}")
plt.axvline(mode_val,   color="purple",ls="dashdot", lw=2, label=f"Mode   = {mode_val:,.0f}")

plt.axvline(mean_val - std_val, color="red", ls="-.", lw=2,
            label=f"-1 SD  = {mean_val - std_val:,.0f}")
plt.axvline(mean_val + std_val, color="red", ls="-.", lw=2,
            label=f"+1 SD  = {mean_val + std_val:,.0f}")

# 5 ▸  ARSIRAN ±1 SD
# ----------------------------------------------------------------
plt.axvspan(mean_val - std_val, mean_val + std_val,
            color="red", alpha=0.10, label="±1 SD range")

# 6 ▸  TATA LETAK
# ----------------------------------------------------------------
plt.title("Distribusi SalePrice dengan Mean, Median, Mode & ±1 SD")
plt.xlabel("SalePrice (USD)")
plt.ylabel("Frekuensi")
plt.legend()
plt.tight_layout()
plt.show()
```
